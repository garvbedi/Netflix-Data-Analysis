```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.chdir(r"C:\Users\haris\Downloads\project\GARV\Netflix-Data-Analysis")
print(os.getcwd())
# Better chart style
sns.set_style("whitegrid")
```

```python
#Dataset preview
df = pd.read_csv(r"C:\Users\haris\Downloads\project\GARV\Netflix-Data-Analysis\netflix_titles.csv")
df.head()
```

```python
#Dataset Information
df.info()
```

```python
#Missing Value check
df.isnull().sum()
```

```python
#Data Cleaning
df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Unknown")
df["country"] = df["country"].fillna("Unknown")
df["date_added"] = df["date_added"].fillna("Not Available")
df["rating"] = df["rating"].fillna("Not Rated")
df["duration"] = df["duration"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

df["type"].value_counts().plot(kind="bar")
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.savefig(r"C:\Users\haris\Downloads\project\GARV\Netflix-Data-Analysis\movies_vs_tvshows.png", bbox_inches="tight")
print("Image saved successfully")
plt.show()
```

```python
# Top Countries
import matplotlib.pyplot as plt

top_countries = df["country"].value_counts().head(10)

top_countries.plot(kind="bar", figsize=(10,5))
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.savefig("top_countries.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df["type"].value_counts()
```

```python
#Release Year Trends
year_data = df["release_year"].value_counts().sort_index()

year_data.plot(figsize=(12,5))
plt.title("Netflix Content Released Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.savefig("release_year_trend.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
df["release_year"].max()
```

```python
#Genre Analysis
top_genres = df["listed_in"].value_counts().head(10)

top_genres.plot(kind="barh", figsize=(10,6))
plt.title("Top Genres on Netflix")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.savefig("top_genres.png", dpi=300, bbox_inches="tight")
plt.show()
```
